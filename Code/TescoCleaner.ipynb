```python
# Load your CSV file
file_path = "/content/drive/MyDrive/tesco-urls/final/semicleaned_file.csv"
df = pd.read_csv(file_path, low_memory=False)

# Identify columns containing "energy" (case-insensitive)
energy_columns = [col for col in df.columns if "energy" in col.lower()]

# Combine the columns into a new "Merged_Energy" column
df["Merged_Energy"] = df[energy_columns].apply(
    lambda row: ", ".join(row.dropna().astype(str).str.strip()), axis=1
)

# Optionally drop the original "energy" columns
df = df.drop(columns=energy_columns)


# fat columns
fat_columns = [col for col in df.columns if "fat" in col.lower()]
df["Merged_Fat"] = df[fat_columns].apply(
    lambda row: ", ".join(row.dropna().astype(str).str.strip()), axis=1
)
df = df.drop(columns=fat_columns)

# protein columns
protein_columns = [col for col in df.columns if "protein" in col.lower()]
df["Merged_Protein"] = df[protein_columns].apply(
    lambda row: ", ".join(row.dropna().astype(str).str.strip()), axis=1
)
df = df.drop(columns=protein_columns)

# carbohydrate columns
carb_columns = [col for col in df.columns if "carb" in col.lower()]
df["Merged_Carbs"] = df[carb_columns].apply(
    lambda row: ", ".join(row.dropna().astype(str).str.strip()), axis=1
)
df = df.drop(columns=carb_columns)

# fiber columns
fiber_columns = [col for col in df.columns if "fib" in col.lower()]
df["Merged_Carbs"] = df[fiber_columns].apply(
    lambda row: ", ".join(row.dropna().astype(str).str.strip()), axis=1
)
df = df.drop(columns=fiber_columns)
```
